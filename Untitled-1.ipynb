```python
# 2025 yılı verilerini filtreleyelim
df_2025 = df[df['Date'].dt.year == 2025]

df_2025
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='variable', y='value', data=pd.melt(df[['Open', 'High', 'Low', 'Close']]))
plt.title("Hisse Senedi Fiyatlarının Kutu Grafiği")
plt.ylabel("Fiyat")
plt.xlabel("Fiyat Türü")
plt.grid(True, alpha=0.3)

# Aykırı değerleri tespit etmek için istatistiksel hesaplamalar
for column in ['Open', 'High', 'Low', 'Close']:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    alt_sinir = Q1 - 1.5 * IQR
    ust_sinir = Q3 + 1.5 * IQR
    aykiri_degerler = df[(df[column] < alt_sinir) | (df[column] > ust_sinir)][column]
    print(f"\n{column} için aykırı değerler:")
    print(f"Alt sınır: {alt_sinir:.2f}")
    print(f"Üst sınır: {ust_sinir:.2f}")
    print(f"Aykırı değer sayısı: {len(aykiri_degerler)}")

plt.show()
```

```python
df['Year'] = df['Date'].dt.year
yillik_degisim = df[(df['Year'] >= 2000) & (df['Year'] < 2025)].groupby('Year')['Close'].agg(['first', 'last'])
yillik_degisim
```

```python
yillik_degisim['Yillik_Degisim_Yuzdesi'] = ((yillik_degisim['last'] - yillik_degisim['first']) / yillik_degisim['first'] * 100)
```

```python
yillik_degisim['Yillik_Degisim_Yuzdesi']
```

```python
# Yıllık değişimleri görselleştirelim

plt.figure(figsize=(40,20))
plt.bar(yillik_degisim.index, yillik_degisim['Yillik_Degisim_Yuzdesi'])
plt.title('Walmart Hisse Senedi Yıllık Değişim Yüzdeleri')
plt.xlabel('Yıl')
plt.ylabel('Değişim Yüzdesi (%)')
plt.grid(True)
plt.xticks(rotation=45)

# Pozitif ve negatif değişimleri farklı renklerle gösterelim ve değerleri ekleyelim
for i, v in enumerate(yillik_degisim['Yillik_Degisim_Yuzdesi']):
    color = 'green' if v >= 0 else 'red'
    plt.bar(yillik_degisim.index[i], v, color=color)
    plt.text(yillik_degisim.index[i], v, f'{yillik_degisim.index[i]}\n{v:.1f}%', 
             ha='center', va='bottom' if v >= 0 else 'top')

plt.show()

print("\nYıllık Değişim İstatistikleri:")
print(yillik_degisim.round(2))
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Tarihi index olarak ayarlayalım
df.set_index('Date', inplace=True)

# Mevsimsel ayrıştırma yapılıyor
decomposition = seasonal_decompose(df['Close'], period=252)  # 252 işlem günü (1 yıl)

# Plotly ile interaktif grafik oluşturma
fig = make_subplots(rows=4, cols=1, subplot_titles=('Orijinal Veri', 'Trend', 'Mevsimsel', 'Artık'))

# Orijinal veri
fig.add_trace(
    go.Scatter(x=df.index, y=df['Close'], name='Orijinal',
               hovertemplate='Tarih: %{x}<br>Değer: %{y:.2f}<extra></extra>'),
    row=1, col=1
)

# Trend
fig.add_trace(
    go.Scatter(x=df.index, y=decomposition.trend, name='Trend',
               hovertemplate='Tarih: %{x}<br>Trend: %{y:.2f}<extra></extra>'),
    row=2, col=1
)

# Mevsimsel
fig.add_trace(
    go.Scatter(x=df.index, y=decomposition.seasonal, name='Mevsimsel',
               hovertemplate='Tarih: %{x}<br>Mevsimsel: %{y:.2f}<extra></extra>'),
    row=3, col=1
)

# Artık (Residual)
fig.add_trace(
    go.Scatter(x=df.index, y=decomposition.resid, name='Artık',
               hovertemplate='Tarih: %{x}<br>Artık: %{y:.2f}<extra></extra>'),
    row=4, col=1
)

# Grafik düzeni güncelleme
fig.update_layout(
    height=1000,
    title_text='Hisse Senedi Fiyat Mevsimsel Ayrıştırma Analizi',
    showlegend=True,
    hovermode='x unified'
)

# X ekseni başlıkları
fig.update_xaxes(title_text='Tarih', row=4, col=1)

# Y ekseni başlıkları
fig.update_yaxes(title_text='Fiyat', row=1, col=1)
fig.update_yaxes(title_text='Trend', row=2, col=1)
fig.update_yaxes(title_text='Mevsimsel', row=3, col=1)
fig.update_yaxes(title_text='Artık', row=4, col=1)

fig.show()
```

```python
import plotly.graph_objects as go
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose

# Tarihi index olarak ayarla ve sırala



# Mevsimsel ayrıştırma yap
decomposition = seasonal_decompose(df['Close'], period=252)  # 252 işlem günü (1 yıl)
seasonal = decomposition.seasonal

# Aylık mevsimsel etkileri hesapla
df['Ay'] = df.index.month
monthly_seasonal = seasonal.groupby(df['Ay']).mean()

# Ay isimleri
ay_isimleri = {
    1: 'Ocak', 2: 'Şubat', 3: 'Mart', 4: 'Nisan', 
    5: 'Mayıs', 6: 'Haziran', 7: 'Temmuz', 8: 'Ağustos',
    9: 'Eylül', 10: 'Ekim', 11: 'Kasım', 12: 'Aralık'
}

# Plotly ile bar grafiği
fig = go.Figure()

# Bar grafiği
fig.add_trace(go.Bar(
    x=list(ay_isimleri.values()),
    y=monthly_seasonal.values,
    name='Mevsimsel Etki',
    marker_color=['red' if x < 0 else 'green' for x in monthly_seasonal.values],
    hovertemplate='Ay: %{x}<br>Mevsimsel Etki: $%{y:.2f}<extra></extra>'
))

# Sıfır çizgisi
fig.add_hline(y=0, line_dash="dash", line_color="gray")

fig.update_layout(
    title={
        'text': 'Walmart Hissesi Aylık Mevsimsel Etki Analizi',
        'y':0.95,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top',
        'font': dict(size=20)
    },
    xaxis_title='Aylar',
    yaxis_title='Mevsimsel Etki ($)',
    hovermode='x unified',
    yaxis=dict(
        zeroline=True,
        zerolinewidth=2,
        zerolinecolor='gray'
    ),
    showlegend=False,
    width=1000,
    height=600,
    bargap=0.2
)

fig.show()
```
